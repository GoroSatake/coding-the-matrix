## p.44 課題0.6.6、課題0.6.7、課題0.6.8

```python
def makeInverseIndex(strlist):
    dic = {}
    for (i, doc) in enumerate(strlist):
        for s in doc.split():
            if s not in dic:
                dic[s] = {i}
            else:
                dic[s].add(i)
    return dic  
```

```python
def orSearch(inverseIndex, query):
    s = set()
    for word in query:
        s |= inverseIndex[word]
    return s
```

```python
def andSearch(inverseIndex, query):
    s = inverseIndex[query[0]]
    for word in query[1:]:
        s &= inverseIndex[word]
    return s
```

```python
f = open('stories_small.txt')
strlist = list(f)
```

```python
inverseIndex = makeInverseIndex(strlist)
```

```python
orSearch(inverseIndex, ['prep', 'course'])
```

```python
andSearch(inverseIndex, ['prep', 'course'])
```
